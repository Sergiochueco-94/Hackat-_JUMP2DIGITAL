# Funciones

```python
# Creamos una función simple para hacer un report simple sobre los DataFrame suministrados
def report_df(dataframe, verbose = True):
    print("El tamaño del dataset es", dataframe.shape[0], "filas y", dataframe.shape[1], "columnas")
    print()
    print(dataframe.info(verbose = verbose))
    total_nulos = dataframe.isnull().sum().sum()
    print()
    print(f"Tenemos un total de {total_nulos} nulos")
```

```python
# Función para convertir columnas categoricas a numéricas de cara a preprocesar los datos y preparlos para un modelo
def OHE (dataframe, column_name):
    _dummy_dataset = pd.get_dummies(dataframe[column_name],prefix=column_name)
    dataframe = pd.concat([dataframe,_dummy_dataset],axis=1)
    return dataframe.drop(column_name,axis=1)
```

#### Para poner todos los precios en la misma escala, es decir "Lloguer mitjà mensual (Euros/mes)" o "Lloguer mitjà per superfície (Euros/m2 mes)", he buscado otro archivo donde recoge la superficie media de las viviendas en Barcelona. De esta manera, calculamos la media de m2 por vivienda en determinadas zonas y las medidas las podemos multiplicar por el precio del m2 y de esa manera obtener una columna con todos los valores de alquiler en "Precio medio mensual (Euros/mes)".

```python
print(df_lloguer.shape) #Para comprobar que no hace un cartesiano medimos el tamaño antes y después

df_lloguer = pd.merge(
    df_lloguer,
    df_superficie,
    how='left',
    left_on = ["Codi_Districte","Nom_Districte","Codi_Barri","Nom_Barri"],
    right_on= ["Codi_districte","Nom_districte","Codi_barri","Nom_barri"],
    )

print(df_lloguer.shape) #Para comprobar que no hace un cartesiano medimos el tamaño antes y después
```

```python
columnas_eliminar = ["Codi_districte","Nom_districte","Codi_barri","Nom_barri","Any"]
```

```python
df_lloguer.drop(columnas_eliminar,axis=1,inplace=True)
```

```python
df_lloguer.head(3)
```

#### Creamos una columna nueva donde agruparemos todos los precios en un mismo formato , precio_medio_mensual. De esta manera no tendremos algunos precios por m2 y otros por apartamento entero

```python
df_lloguer['Preu_mitja_mensual'] = pd.Series()
```

#### Separamos en 2 el DataFrame, según los valores de Lloguer_mitja, que es la columna donde se hace la división que comentabamos anteriormente

```python
df_lloguer_mitja_mensual = df_lloguer[df_lloguer['Lloguer_mitja']=="Lloguer mitjà mensual (Euros/mes)"]
df_lloguer_mitja_superficie = df_lloguer[df_lloguer['Lloguer_mitja']=="Lloguer mitjà per superfície (Euros/m2 mes)"]
```

#### Calculamos para los valores donde el Lloguer_mitja == m2, multiplicamos las columnas precio y superficie para obtener el precio medio mensual del alquiler

#### Y para el resto simplemente lo multiplicamos por 1 para obtener el mismo valor que ya tenia 

```python
df_lloguer_mitja_superficie['Preu_mitja_mensual'] = df_lloguer_mitja_superficie['Preu']*df_lloguer_mitja_superficie['Superficie_mitjana_habitatge_(m2)']
df_lloguer_mitja_mensual['Preu_mitja_mensual'] = df_lloguer_mitja_mensual['Preu']*1
```

#### Una vez hecho los cálculos concatenamos los dos sub DataFrame en uno solo para volver a obtener el original con el cálculo hecho en la nueva columna. 

```python
df_lloguer = pd.concat([df_lloguer_mitja_mensual,df_lloguer_mitja_superficie])
```

# Imputación de valores nulos en precio

```python
df_lloguer.isnull().sum() #Seguimos teniendo la misma cantidad a pesar de los cambios, así que es buena señal
```

#### Usaremos el KNNImputer para imputar los precios en función de sus vecinos más cercanos, de esta manera podemos imputar nulos y el agoritmo buscará aquellos más parecidos (barrio, distrito,etc), para imputar valores 

```python
imputer = KNNImputer()
```

```python
df_lloguer.describe(include = np.number).columns
```

```python
df_lloguer_num = df_lloguer[
    ['Trimestre',
    'Codi_Districte', 
    'Codi_Barri',
    'Superficie_mitjana_habitatge_(m2)', 
    'Preu_mitja_mensual']]
```

```python
df_lloguer_num = imputer.fit_transform(df_lloguer_num)
```

```python
df_lloguer_num.isnull().sum() #Ya hemos imputado los valores nulos
```

#### Vamos a concatenar las otras columnas que no hemos seleccionado antes para poder concatenar con el df_poblacio

#### Como no hemos alterado el orden del dataset podemos hacer un concat en axis=1, podemos estar seguros de que seguirá el mismo orden que debería

```python
df_lloguer_imputed_nulos = pd.concat([df_lloguer_num,df_lloguer[['Nom_Districte','Nom_Barri']]],axis=1)
```
